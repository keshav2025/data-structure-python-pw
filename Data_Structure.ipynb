#  What are the main differences between lists and tuples in Python ?
lists and tuples are both used to store collections of items, but they have distinct characteristics that make them suitable for different scenarios.

list are mutable collections, allowing changes like adding or removing items, and are defined using square brackets []. They are ideal for dynamic data that may change over time. In contrast, tuples are immutable, meaning their contents cannot be altered after creation, and are defined using parentheses (). Tuples are suitable for fixed data that should remain constant, offering faster performance and lower memory usage compared to lists.

#  Describe how dictionaries store data

dictionaries store data as key-value pairs using a data structure called a hash table. This allows for efficient data retrieval, insertion, and deletion operations. Key-Value Pairs Each item in a dictionary consists of a unique key and its associated value. Keys must be immutable types (like strings, numbers, or tuples) ensuring they remain constant and hashable.


# Why might you use a set instead of a list in Python

When to Use a Set
Eliminating Duplicates --> Sets automatically discard duplicate entries, making them ideal for ensuring all elements are unique.
Fast Membership Testing --> Checking if an item exists in a set is generally faster (average-case O(1)) compared to a list (O(n)), especially beneficial for large datasets.

When to Use a List

Ordered Data --> Lists maintain the order of elements, which is essential when the sequence matters

Allowing Duplicates -->Lists can contain duplicate elements, suitable for scenarios where repetition is necessary

Indexing and Slicing --> Lists support accessing elements by index and slicing, offering flexibility in data manipulation.


# What is a string in Python, and how is it different from a list ?

​In Python, a string is a sequence of characters enclosed in quotes, such as "Hello". Strings are immutable, meaning their contents cannot be changed after creation. A list is an ordered collection of items enclosed in square brackets, like [1, 2, 3]. Lists are mutable, allowing you to modify their contents by adding, removing, or changing elements.
